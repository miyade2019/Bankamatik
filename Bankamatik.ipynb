```python
# Bankamatik
# ilk başlangıçta hiç hesap yok
# en az iki tane hesap açılacak
# para çekme
# para yatırma
# bakiye sorma
# kullanıcı adı ve şifre sorsun

menu="""
1- Bakiye sorma
2- Hesap oluşturma
3- Para çekme
4- Para yatırma
5- Çıkış
"""


hesaplar=[]

k_adi="asli"
sifre="12345"

u_name=input("Kullanıcı adını giriniz: ")
u_sifre=input("Şifreyi giriniz: ")

kontrol=False

if u_name == k_adi and u_sifre == sifre:

  kontrol=True

  if len(hesaplar)==0:

    print("Hiç hesabınız yok, yeni bir hesap oluşturunuz:")
    miktar=int(input("Hesabınızı kaç lira ile açmak istersiniz: "))

    hesaplar.append(miktar)

else:
  print("Kullanıcı adı veya şifre yanlış")



# Kullanıcı adı ve şifre doğruysa döngüye gir.
if kontrol==True:

  while True:


    # kullanıcı adı ve şifre doğruysa..



      print(menu)

      sec=input("Seçiminiz: ")

      #çıkış
      if sec == "5":
        print("Hoşçakalın")
        break

      #bakiye sorma
      elif sec == "1":
        for i in range(len(hesaplar)):
          print(i+1,". hesabınız:",hesaplar[i])

      #hesap oluşturma
      elif sec == "2":
        miktar=int(input("Hesabınızı kaç lira ile açmak istersiniz: "))
        hesaplar.append(miktar)
        print("Hesabınız oluşturuldu.")

      #para çekme
      elif sec == "3":
        for i in range(len(hesaplar)):
            print(i+1,". hesabınız:",hesaplar[i])

        hesap_no=int(input("Hangi hesaptan para çekmek istiyorsunuz: ")) - 1
        c_miktar = int(input("Kaç lira çekeceksiniz: "))

        if c_miktar > hesaplar[hesap_no]:
          print("Bakiye yetersiz.")
        else:
          hesaplar[hesap_no] -= c_miktar


      #para yatırma
      elif sec == "4":
        for i in range(len(hesaplar)):
            print(i+1,". hesabınız:",hesaplar[i])

        hesap_no=int(input("Hangi hesaba para yatırmak istiyorsunuz: ")) - 1
        c_miktar = int(input("Kaç lira yatıracaksınız: "))

        hesaplar[hesap_no] += c_miktar
```
